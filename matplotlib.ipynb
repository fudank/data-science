```python
from matplotlib import pyplot as plt

a = [5,7,8,3,7,9]
x = [1,2,3,4,5,6]
y = [5,6,7,8,9,5]
x1 = [7,9,8,3,2,22]
y1 = [2,7,12,2,23,3]

plt.plot([],[],color = 'm',label = 'x',linewidth = 5)

plt.plot([],[],color = 'c',label = 'y',linewidth = 5)

plt.plot([],[],color = 'r',label = 'x1',linewidth = 5)

plt.plot([],[],color = 'k',label = 'y1',linewidth = 5)
plt.stackplot(a,x,y,x1,y1,colors = ['m','c','r','k'])



plt.legend()
plt.grid(True,color = 'k')
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("area plot")
plt.show()
```
